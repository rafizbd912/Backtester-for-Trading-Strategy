```python
import io
import pandas as pd
import numpy as np
from google.colab import files

# Upload CSV files
uploaded = files.upload()
```

```python
# Read CSV data from memory
dfs = {}
for filename in uploaded.keys():
    data = uploaded[filename].decode('utf-8')
    df = pd.read_csv(io.StringIO(data))
    df['Date'] = pd.to_datetime(df['Date'])
    dfs[filename[:-4]] = df  # Store dataframes in a dictionary using ticker symbols as keys
    
# Define a trading strategy based on Bollinger Bands
def strategy(df):
    # Buy if the price is below the lower Bollinger Band
    if df['Close'] < df['Lower']:
        return 1
    # Sell if the price is above the upper Bollinger Band
    elif df['Close'] > df['Upper']:
        return -1
    # Do nothing if the price is between the upper and lower Bollinger Bands
    else:
        return 0

# Calculate returns for each stock
cumulative_returns = []
for ticker, df in dfs.items():
    # Calculate the rolling mean and standard deviation
    df['SMA_20'] = df['Close'].rolling(window=20).mean()
    df['STD_20'] = df['Close'].rolling(window=20).std()

    # Calculate the upper and lower Bollinger Bands
    df['Upper'] = df['SMA_20'] + 2 * df['STD_20']
    df['Lower'] = df['SMA_20'] - 2 * df['STD_20']

    # Apply the trading strategy to the data
    df['signal'] = df.apply(strategy, axis=1)

    # Calculate daily returns based on the signal
    df['returns'] = df['signal'].shift(1) * df['Close'].pct_change()

    # Calculate cumulative returns
    df['cum_returns'] = (1 + df['returns']).cumprod()

    # Print the final cumulative return for the stock
    final_return = df['cum_returns'].iloc[-1]
    print(f'{ticker}: Final cumulative return: {final_return}')
    cumulative_returns.append(final_return)

# Calculate and print the average cumulative return as a percentage
average_return = np.mean(cumulative_returns) - 1
average_return_percent = average_return * 100
print(f'Average percentage return on investment: {average_return_percent:.2f}%')
```
